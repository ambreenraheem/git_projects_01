## **It is very difficult to understand the histograph of each variance,so here we will take help from boxplot to understand the actual Shapiro Wilk test of penguins statistics.**

### Now here you can see the boxplot and shapiro test,both are telling the data of all species veriable(flipper_length_mm) is normal.

```python
plt.title("BOXPLOT OF PENGUINS")
sns.boxplot(x='species',
            y='flipper_length_mm',
            hue='species',
            data=pg)
```

```python
from scipy.stats import shapiro
data=[penguins]
stat,p=shapiro(pg['flipper_length_mm'])
print('stat=',stat)
print('p=',p)
if p > 0.05:
    print("The data is normal.")
else:
    print("The data is not normal.")
```

### This is also normal..

```python
plt.title("BOXPLOT OF PENGUINS")
sns.boxplot(x='species',
            y='bill_length_mm',
            hue='species',
            data=pg)
```
